```
import geopandas
import descartes
import matplotlib.pyplot as plt
import pandas as pd
from shapely.geometry import Point, Polygon
import folium

sd_traffic = pd.read_csv('sd_traffic_2015_stations.csv')
sd_traffic['longitude'] = sd_traffic['longitude']*-1
sd_traffic = sd_traffic[['latitude', 'longitude']]

sd_traffic.head()
locationlist = sd_traffic.values.tolist()
print(locationlist)

map = folium.Map(location=[32.732428,	-117.111704], tiles='Stamen Toner', zoom_start=12, control_scale=True)
marker_cluster = folium.MarkerCluster().add_to(map)

for point in range(0, len(locationlist)):
  folium.Marker(locationlist[point], popup='SD Traffic Stations 2015').add_to(marker_cluster)

# Filepath to the output
outfp = "sd_traffic_map.html"

# Save the map
map.save(outfp)

map  

#geometry = [Point(xy) for xy in zip(sd_traffic['latitude'], sd_traffic['longitude']*-1)]
#geo_df = geopandas.GeoDataFrame(sd_traffic, geometry = geometry) 
#geo_df = geo_df.drop(['latitude', 'longitude'], axis=1)
#geo_df.head()
```
